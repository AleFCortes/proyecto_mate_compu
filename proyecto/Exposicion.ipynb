## 2. Partícula por partícula...

Comenzamos simulando el movimiento de una partícula al aplicarle una fuerza dada durante un tiempo específico. A partir de la fuerza y de la masa del objeto, podemos calcular la magnitud con la que acelera a través de la **Segunda ley de Newton**:

$$
F = m a
$$

Una vez determinada la aceleración podemos aplicar las fórmulas vistas en secundaria:

$$
v_f = v_i t  + a t
$$

$$
\Delta d = v_i t + \frac{1}{2} a t^2
$$
