## 1. Introducción

Los cuerpos interactúan unos con otros y su movimiento depende de su posición en el espacio y de sus masas. El problema de cómo se relacionan *dos* cuerpos ha sido resuelto. No obstante, para tres o más cuerpos el problema presenta diversas complicaciones.

En esta simulación nos basaremos en la resolución del problema sobre dos cuerpos para aplicarlo a más objetos.

## 2. Partícula por partícula...

Comenzamos simulando el movimiento de una partícula al aplicarle una fuerza dada durante un tiempo específico. A partir de la fuerza y de la masa del objeto, podemos calcular la magnitud con la que acelera a través de la **Segunda ley de Newton**:

$$
F = m a
$$

Una vez determinada la aceleración podemos aplicar las fórmulas vistas en secundaria:

$$
v_f = v_i t  + a t
$$

$$
d_f = v_i t + \frac{1}{2} a t^2 + d_i
$$

El código está en el archivo `class_particula.py` .

```python
from class_particula import Particula
```

Simulando el movimiento de una partícula:

```python
posicion_inicial = np.array([0.0, 0.0])
velocidad_inicial = np.array([-780.0, 33.3])
masa_inicial = 10
fuerza_inicial = np.array([200.0, -50.0])
time = 83.0
steps = 1000 #Entero
```

```python
particula_1 = Particula(posicion = posicion_inicial, velocidad = velocidad_inicial, masa = masa_inicial)
```

```python
particula_1.mover(fuerza = fuerza_inicial, tiempo = time, pasos = steps)
```

```python
particula_1.dibujar()
```

Para visualizar mejor la trayectoria de la partícula a lo largo de los ejes X, Y, definimos el siguient método:

```python
particula_1.dibujar_xt_yt()
```

¿En qué nos ayuda steps?

```python
def moviendo_pasos(f, t, p):
    """
    f es la fuerza, t el tiempo y p los pasos
    """
    particula_1.mover(fuerza = f, tiempo = t, pasos = p)
    particula_1.dibujar_xt_yt()
```

```python
particula_1 = Particula(posicion = posicion_inicial, velocidad = velocidad_inicial, masa = masa_inicial)
```

```python
interact(moviendo_pasos, f = fixed(fuerza_inicial), t = fixed(time), p = (0,100))
```

Jugando con la fuerza...

```python
posicion_inicial = np.array([0.0, 0.0])
velocidad_inicial = np.array([-20.0, 50.0])
masa_inicial = 10
fuerza_inicial = np.array([-20.0, -50.0])
time = 83.0
steps = 1000
```

```python
particula_2 = Particula(posicion = posicion_inicial, velocidad = velocidad_inicial, masa = masa_inicial)
```

```python
def cambiar_fuerza(f_x, f_y, t, p):
    """
    f_x es la fuerza en x
    f_y es la fuerza en y
    t es el tiempo
    p son los pasos
    """
    f = np.array([f_x, f_y])
    particula_2.mover(fuerza = f, tiempo = t, pasos = p)
    particula_2.dibujar_xt_yt()
```

```python
interact(cambiar_fuerza, f_x = (-100.0, 100.0), f_y = (-100.0, 100.0), t = fixed(time), p = fixed(steps))
```

## 3. ¿Más fuerzas?

A partir de la **Ley de Gravitación Universal** que le debemos a Newton,

$$
F = \frac{G m_1 m_2}{r^2}
$$ 

podemos determinar la fuerza que ejerce una partícula sobre otra. Si existen más de dos partículas, la fuerza total que se ejerce en cada una de ellas, por el _Principio de superposición_, es la suma de la fuerzas que se ejercen particularmente.

De aquí que nuestra _Particula_ recibirá muchas fuerzas en el espacio: entonces ¿cómo modelar que una sola _Particula_ reciba más de una fuerza?

Para resolver este problema, implementamos en _Particula_ el método:

```python
posicion_inicial = np.array([0.0, 0.0])
velocidad_inicial = np.array([-20.0, 50.0])
masa_inicial = 10
```

```python
particula_3 = Particula(posicion = posicion_inicial, velocidad = velocidad_inicial, masa = masa_inicial)
```

```python
forces = [[-34.4,100.0], [23.65,-300.4], [2000.0,-10.0], [-30.0,80.0], [-2.90, 17.0], [200.0, -200.0]] # El método recibe una lista de fuerzas
```

```python
times = [[0.0, 10.0], [10.0, 20.0], [20.0, 30.0], [30.0, 40.0], [40.0, 50.0], [50.0, 60.0]]
```

```python
steps = [100, 100, 100, 100, 100, 100]
```

```python
particula_3.mover_muchas_fuerzas(tiempos = times, fuerzas = forces, paso = steps)
#particula_3.pos_x[len(forces)-1][]
```

```python
particula_3.dibujar_muchas_fuerzas()
```
